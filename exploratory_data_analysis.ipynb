# 3. Exploratory Data Analysis

Analyzing default patterns (PD), recovery potential (LGD), and exposure (EAD).

## PD: Default Patterns
We analyze how default rates vary across different categorical features like Region.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Loan_Default.csv')
# Simplified default rate plot
sns.countplot(x='Region', hue='Status', data=df)
plt.title('Count of Defaults by Region')
plt.show()
```

## LGD and EAD: Exposure & Value
Relationship between loan amount and property value.

```
sns.jointplot(x='property_value', y='loan_amount', data=df[df['Status']==1], kind='hex')
plt.suptitle('Property Value vs Loan Amount for Defaults', y=1.02)
plt.show()
```
