# 1. Dataset Definition and Understanding

In this section, we load the Loan Default dataset to understand its structure and define the variables relevant for credit risk modeling (PD, LGD, EAD).

```
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('/content/Loan_Default.csv')

# Display basic information
print("Dataset Shape:", df.shape)
display(df.head())
display(df.info())
```

### Variable Definitions and Structure

Based on the initial inspection, we can categorize the variables:

*   **Target Variables:**
    *   **Probability of Default (PD):** Typically represented by the `Status` column (0 for no default, 1 for default).
    *   **Loss Given Default (LGD):** Usually derived from recovery columns or by calculating the percentage of loss relative to the exposure at the time of default.
    *   **Exposure at Default (EAD):** Often represented by the total amount outstanding at the time of default, such as `loan_amount` or `property_value`.

*   **Feature Categories:**
    *   **Borrower Information:** Age, income, credit score, etc.
    *   **Loan Characteristics:** Loan type, term, interest rate, loan-to-value (LTV) ratio.
    *   **Property Information:** Property value, occupancy type.

### Dataset Structure
The dataset consists of both numerical and categorical features. Missing values are present in several columns (as seen in `df.info()`), which will require cleaning before modeling.
