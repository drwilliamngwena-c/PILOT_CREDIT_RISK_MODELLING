# 1. Dataset Definition and Understanding

In this section, we load the Loan Default dataset to understand its structure and define the variables relevant for credit risk modeling (PD, LGD, EAD).

```
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('/content/Loan_Default.csv')

# Display basic information and first 5 rows
print("Dataset Shape:", df.shape)
display(df.head())
display(df.info())
```

### Variable Definitions and Structure

Based on the initial inspection, we can categorize the variables:

*   **Target Variables:**
    *   **Probability of Default (PD):** Represented by the `Status` column (0 for no default, 1 for default).
    *   **Loss Given Default (LGD):** This will be calculated in subsequent steps, typically using recovery values or exposure loss ratios.
    *   **Exposure at Default (EAD):** Identified through columns like `loan_amount` or `property_value`.

*   **Feature Categories:**
    *   **Borrower Information:** Age, income, credit score, gender, etc.
    *   **Loan Characteristics:** Loan type, term, interest rate, loan-to-value (LTV) ratio, interest rate spread.
    *   **Property Information:** Property value, occupancy type, total units.

### Dataset Structure
The dataset consists of 34 columns and over 50,000 records. It includes a mix of categorical (object) and numerical (float/int) data types. There is significant missing data in columns like `rate_of_interest` and `Interest_rate_spread`, requiring preprocessing.
