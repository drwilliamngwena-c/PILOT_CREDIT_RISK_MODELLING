# Project README: Credit Risk Data Preprocessing

This notebook provides a visual summary of the data quality and preprocessing steps taken for the Loan Default dataset.

## 1. Outlier Analysis of Key Drivers

We identified **Income**, **Loan Amount**, and **Property Value** as critical variables.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Re-loading data for visual demonstration
df = pd.read_csv('/content/Loan_Default.csv')
key_drivers = ['income', 'loan_amount', 'property_value']
for col in key_drivers: df[col] = pd.to_numeric(df[col], errors='coerce').fillna(df[col].median())

plt.figure(figsize=(15, 5))
for i, col in enumerate(key_drivers):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=df[col])
    plt.title(f'Original: {col}')
plt.tight_layout()
plt.show()
```

## 2. Post-Treatment: Winsorization (Capping)

```
# Applying capping for visualization
for col in key_drivers:
    df[col] = df[col].clip(df[col].quantile(0.01), df[col].quantile(0.99))

plt.figure(figsize=(15, 5))
for i, col in enumerate(key_drivers):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(f'Capped: {col}')
plt.tight_layout()
plt.show()
```
