# 2. Data Preprocessing

In this section, we address data quality issues including missing values and outliers, and prepare the features for modeling.

```
# Missing Values Imputation
numerical_cols = ['income', 'property_value', 'LTV', 'dtir1', 'rate_of_interest', 'Upfront_charges']
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(df[col].median())

cat_cols_to_fix = ['loan_limit', 'approv_in_adv', 'loan_purpose', 'Neg_ammortization']
for col in cat_cols_to_fix:
    df[col] = df[col].fillna(df[col].mode()[0])
print('Missing values imputed.')
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
object_cols = df.select_dtypes(include=['object']).columns
for col in object_cols:
    if col not in ['ID', 'year']:
        df[col] = df[col].astype(str)
        df[col] = le.fit_transform(df[col])
print('Categorical variables encoded.')
```

### 2.1 Outlier Treatment and Visualization

```
import matplotlib.pyplot as plt
import seaborn as sns

key_drivers = ['income', 'loan_amount', 'property_value']

def treat_outliers(series):
    return series.clip(series.quantile(0.01), series.quantile(0.99))

for col in key_drivers:
    df[col] = treat_outliers(df[col])

print('Outliers treated by capping.')
```
