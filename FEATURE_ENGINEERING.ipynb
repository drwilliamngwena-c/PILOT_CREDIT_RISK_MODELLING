# 4. Feature Engineering

Creating predictive indicators for PD, LGD, and EAD modeling.

```
# Engineering new risk drivers
df['loan_to_income'] = df['loan_amount'] / (df['income'] + 1)
df['property_equity_estimate'] = df['property_value'] - df['loan_amount']
df['is_underwater'] = (df['loan_to_income'] > 1).astype(int)
df['Interest_rate_spread'] = pd.to_numeric(df['Interest_rate_spread'], errors='coerce').fillna(0)
df['potential_exposure_risk'] = df['loan_amount'] * (1 + df['Interest_rate_spread'].astype(float) / 100)
```
